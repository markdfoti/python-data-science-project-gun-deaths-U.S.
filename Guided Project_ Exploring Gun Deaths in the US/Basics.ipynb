```python
import datetime
```

```python
dates = []
```

```python
for info in data:
    int_year = int(info[1])
    int_month = int(info[2])
    date = datetime.datetime(year = int_year, month = int_month, day = 1)
    dates.append(date)
```

```python
dates[0:5]
```

```python
date_counts = {}
```

```python
for date in dates:
    if date not in date_counts:
        date_counts[date] = 1
    else:
        date_counts[date] += 1
```
